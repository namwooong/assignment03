```python
import numpy as np
import matplotlib.pyplot as plt

num     = 1001
std     = 5 

# x  : x-coordinate data
# y1 : (clean) y-coordinate data 
# y2 : (noisy) y-coordinate data

def fun(x):
 
	# f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x) 			# clean points
y2      = y1 + nn * std		# noisy points

plt.plot(x, y1, 'b.', x, y2, 'k.')
plt.show()
```

make answer function

```python
def answer(x, p, lamb):
    A = np.array([[x1**pi for pi in range(p)] for x1 in x])
    e = np.sqrt(lamb)*np.eye(p)
    A = np.concatenate((A, e), axis=0)
    Q, R = np.linalg.qr(A)
    b = np.concatenate((y1, np.zeros(p)), axis=None)
    sol = np.linalg.solve(R, np.matmul(Q.T, b))
    f = 0
    for i, temp in enumerate(sol):
        f += temp * (x**i)
    return f
```

make loss function

```python
def calLoss(f, real):
    if f.shape != real.shape: 
        return np.inf
    sum = 0
    for (y1, y2) in zip(f, real):
        sum += (y1 - y2)**2
    return sum
```

calculate p and lamds and loss

```python
for i in range(6, 16):
    for j, l in enumerate([1/(2**k) for k in range(4)]):
        f = answer(x, i, l)
       
        plt.xlim(-11,11)
        plt.ylim(-11,11)
        plt.plot(x, f, 'b.', x, y2, 'k.')
        print("p = "+str(i)+" , lambda = "+str(j))
        print("Loss : "+str(calLoss(f, y2)))
        plt.show()
        
        
        
```
